```python
import pandas as pd
import plotly.express as px

## 1. ‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏à‡∏≤‡∏Å‡∏ï‡∏≤‡∏£‡∏≤‡∏á‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• Squad, MP, GD, xGD
data = {
    'Squad': ['Arsenal', 'Bournemouth', 'Tottenham', 'Sunderland', 'Man City', 'Man Utd',
              'Liverpool', 'Aston Villa', 'Chelsea', 'Crystal Palace', 'Brentford',
              'Newcastle Utd', 'Brighton', 'Everton', 'Leeds United', 'Burnley',
              'Fulham', "Nott'm Forest", 'West Ham', 'Wolves'],
    'MP': [9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9],
    'GD': [13, 5, 10, 4, 10, 1, 2, 1, 6, 3, 0, 1, -1, -3, -5, -5, -5, -12, -13, -12],
    'xGD': [2.2, -0.1, 1.8, 0.4, 6.6, 1.8, 3.8, -0.5, 0.0, 0.3, 1.2, 0.0, 1.4, -1.2, -0.2, -1.3, -3.8, -5.5, -7.2, -2.5],
}

df = pd.DataFrame(data)

## 2. ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡πà‡∏≤ Over/Under Performance (OP)
# OP = GD - xGD
df['OP'] = df['GD'] - df['xGD']

## 3. ‡∏à‡∏±‡∏î‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡∏ô‡∏≥‡πÄ‡∏™‡∏ô‡∏≠‡πÑ‡∏î‡πâ‡∏î‡∏µ (‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏à‡∏≤‡∏Å Overperform ‡∏°‡∏≤‡∏Å‡πÑ‡∏õ Underperform ‡∏°‡∏≤‡∏Å)
df = df.sort_values(by='OP', ascending=False)

## 4. ‡∏™‡∏£‡πâ‡∏≤‡∏á Interactive Bar Chart ‡∏î‡πâ‡∏ß‡∏¢ Plotly Express
fig = px.bar(
    df,
    x='OP',
    y='Squad',
    orientation='h',  # ‡∏Å‡∏£‡∏≤‡∏ü‡πÅ‡∏ó‡πà‡∏á‡πÅ‡∏ô‡∏ß‡∏ô‡∏≠‡∏ô
    color='OP',
    color_continuous_scale=px.colors.diverging.RdYlBu, # ‡πÉ‡∏ä‡πâ‡∏™‡∏µ‡πÅ‡∏ö‡∏ö‡πÑ‡∏•‡πà‡πÇ‡∏ó‡∏ô (‡πÅ‡∏î‡∏á-‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡∏á-‡∏ô‡πâ‡∏≥‡πÄ‡∏á‡∏¥‡∏ô)
    title="üí° Over/Under Performance (OP): GD vs xGD Premier League",
    labels={'OP': 'Over/Under Performance (OP = GD - xGD)', 'Squad': 'Team'},
)

# ‡∏õ‡∏£‡∏±‡∏ö‡πÅ‡∏Å‡∏ô x ‡πÉ‡∏´‡πâ‡∏≠‡∏¢‡∏π‡πà‡∏ï‡∏£‡∏á‡∏Å‡∏•‡∏≤‡∏á (‡∏®‡∏π‡∏ô‡∏¢‡πå‡∏Å‡∏•‡∏≤‡∏á‡∏ó‡∏µ‡πà 0)
max_abs = df['OP'].abs().max()
fig.update_layout(xaxis_range=[-max_abs, max_abs])

# ‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÄ‡∏™‡πâ‡∏ô‡πÅ‡∏ô‡∏ß‡∏ï‡∏±‡πâ‡∏á‡∏ó‡∏µ‡πà OP = 0 ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡πÄ‡∏´‡πá‡∏ô‡∏Å‡∏≤‡∏£‡πÅ‡∏ö‡πà‡∏á‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á Overperform ‡πÅ‡∏•‡∏∞ Underperform ‡∏ä‡∏±‡∏î‡πÄ‡∏à‡∏ô
fig.add_vline(x=0, line_width=1, line_dash="dash", line_color="black")

# ‡∏õ‡∏£‡∏±‡∏ö Hover Data (‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡πÅ‡∏™‡∏î‡∏á‡πÄ‡∏°‡∏∑‡πà‡∏≠‡∏ô‡∏≥‡πÄ‡∏°‡∏≤‡∏™‡πå‡πÑ‡∏õ‡∏ä‡∏µ‡πâ)
fig.update_traces(
    hovertemplate="<b>%{y}</b><br>" +
                  "OP (GD - xGD): %{x:.2f}<br>" +
                  "Goal Difference (GD): %{customdata[0]}<br>" +
                  "Expected Goal Difference (xGD): %{customdata[1]:.2f}<extra></extra>",
    customdata=df[['GD', 'xGD']] # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå GD ‡πÅ‡∏•‡∏∞ xGD ‡πÄ‡∏Ç‡πâ‡∏≤‡πÑ‡∏õ‡πÉ‡∏ô customdata
)

# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏Å‡∏£‡∏≤‡∏ü
fig.show()

# ‡∏™‡∏£‡∏∏‡∏õ Insight (‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÉ‡∏ä‡πâ‡∏õ‡∏£‡∏∞‡∏Å‡∏≠‡∏ö‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå)
print("\n--- Insight Summary ---")
print(f"**Overperformers ‡∏ó‡∏µ‡πà‡πÇ‡∏î‡∏î‡πÄ‡∏î‡πà‡∏ô:** ‡∏ó‡∏µ‡∏°‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏Ñ‡πà‡∏≤ OP ‡πÄ‡∏õ‡πá‡∏ô‡∏ö‡∏ß‡∏Å‡∏°‡∏≤‡∏Å‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î")
top_overperform = df.head(3)
print(top_overperform[['Squad', 'OP', 'GD', 'xGD']])

print(f"\n**Underperformers ‡∏ó‡∏µ‡πà‡∏ô‡πà‡∏≤‡∏Å‡∏±‡∏á‡∏ß‡∏•:** ‡∏ó‡∏µ‡∏°‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏Ñ‡πà‡∏≤ OP ‡πÄ‡∏õ‡πá‡∏ô‡∏•‡∏ö‡∏°‡∏≤‡∏Å‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î")
top_underperform = df.tail(3).sort_values(by='OP', ascending=True)
print(top_underperform[['Squad', 'OP', 'GD', 'xGD']])
```
